```python
import cv2
import glob
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
airplane = [cv2.imread(file) for file in glob.glob("images/Airplane/*.png")]
auto = [cv2.imread(file) for file in glob.glob("images/Automobile/*.png")]
bird = [cv2.imread(file) for file in glob.glob("images/Bird/*.png")]
cat = [cv2.imread(file) for file in glob.glob("images/Cat/*.png")]
deer = [cv2.imread(file) for file in glob.glob("images/Deer/*.png")]
dog = [cv2.imread(file) for file in glob.glob("images/Dog/*.png")]
frog = [cv2.imread(file) for file in glob.glob("images/Frog/*.png")]
horse = [cv2.imread(file) for file in glob.glob("images/Horse/*.png")]
ship = [cv2.imread(file) for file in glob.glob("images/Ship/*.png")]
truck = [cv2.imread(file) for file in glob.glob("images/Truck/*.png")]
images = airplane + auto + bird + cat + deer + dog + deer + dog + frog + horse + ship + truck
```

```python
def check(photo, result):
    for img in result:
        if(distance(photo, img) == 0.0):
            return True
```

```python
def distance(imageA, imageB):
    dist = np.sqrt(np.sum((imageA.astype("float") - imageB.astype("float")) ** 2))
    dist /= float(imageA.shape[0] * imageA.shape[1])
    return dist

def calculate_neighbours(testcase):
    base = random.choice(testcase)
    result = [base]
    resulted_distance = []
    for i in range(1, 5):
        min = np.inf
        for photo in images:
            if(check(photo, result) == True):
                continue
            elif(distance(base, photo) <= min):
                temp = photo
                min = distance(base, photo)
        result.append(temp)
        resulted_distance.append(min)
    return result, resulted_distance
```

```python

def print_images(resulted_images):
    fig=plt.figure(figsize=(12, 12))
    columns = 5
    rows = 1
    for i in range(1, columns*rows+1):
        fig.add_subplot(rows, columns, i)
        plt.imshow(resulted_images[i-1])
        plt.axis("off")
    plt.show()
```

```python
test1, test_distance1 = calculate_neighbours(testcase = airplane)
print_images(test1)
#print("Distance: ", test_distance1)
```
